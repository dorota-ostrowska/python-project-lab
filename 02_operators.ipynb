## Operatory

### Operatory arytmetyczne

Pozwalają wykonywać operacje matematyczne jak dodawanie, odejmowanie etc.

Screenshot 2022-12-19 at 17.13.18.png

```python
a = 7
b = 2

# dodawanie
print ('Sum: ', a + b)  

# odejmowanie
print ('Subtraction: ', a - b)   

# mnożenie
print ('Multiplication: ', a * b)  

# dzielenie
print ('Division: ', a / b) 

# reszta z dzielenia
print ('Modulo: ', a % b)  

# a do potęgi b
print ('Power: ', a ** b)  

# dzielenie z zaokrągleniem w dół
print ('Floor division: ', a // b)
```

### Operatory do przypisania wartości do zmiennej

Screenshot 2022-12-19 at 17.16.26.png

```python
# przypisujemy 10 do a
a = 10

# przypisujemy 5 do b
b = 5 

# przypisujemy sumę a i be do a
a += b      # a = a + b

print(a)
```

### Operatory do porównania

Screenshot 2022-12-19 at 17.25.06.png

```python
a = 5

b = 2

# równa się
print('a == b =', a == b)

# nie równa się
print('a != b =', a != b)

# więcej niż
print('a > b =', a > b)

# mniej niż
print('a < b =', a < b)

# większy równy
print('a >= b =', a >= b)

# mniejszy równy
print('a <= b =', a <= b)
```

### Operatory logiczne

Screenshot 2022-12-19 at 17.27.53.png

```python
# Jakie będą wyniki?
# logiczne AND
print(True and True)     
print(True and False)    

# logiczne OR
print(True or False)     

# logiczne NOT
print(not True)          
```

### Operatory specjalne

Screenshot 2022-12-19 at 17.29.25.png

```python
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]
y3 = [1,2,3]

# Jakie będą wyniki?

print(x1 is not y1)  

print(x2 is y2)  

print(x3 is y3)  
```

### Operatory przynależności

Screenshot 2022-12-19 at 17.31.07.png

```python
message = 'Hello world'
dict1 = {1:'a', 2:'b'}

# sprawdź czy 'H' występuje w stringu message
print('H' in message)  

# sprawdź czy 'hello' występuje  stringu message
print('hello' not in message)  

# sprawdź czy klucz '1' występuje w słowniku dict1
print(1 in dict1)  

# sprawdź czy klucz 'a' występuje w słowniku dict1
print('a' in dict1) 
```
